```python
# Limpiar y convertir tipos
try:
    # Convertir fechas con UTC=True para manejar zonas horarias mixtas
    for col in ['created', 'modified']:
        if df[col].notna().any():  # Verificar si hay datos no nulos
            df[col] = pd.to_datetime(df[col], errors='coerce', utc=True)
    
    # Convertir start_date y end_date solo si tienen datos
    for col in ['start_date', 'end_date']:
        if df[col].notna().any():
            df[col] = pd.to_datetime(df[col], errors='coerce', utc=True)
    
    # Convertir booleanos verificando primero el tipo
    for col in ['addresses_main_address', 'values_outstanding']:
        if df[col].dtype == object:  # Solo convertir si es object
            df[col] = df[col].map({'True': True, 'False': False})
    
    # Convertir códigos postales a string de manera segura
    df['addresses_zip_code'] = df['addresses_zip_code'].astype(str)
    df.loc[df['addresses_zip_code'] != 'nan', 'addresses_zip_code'] = \
        df.loc[df['addresses_zip_code'] != 'nan', 'addresses_zip_code'].str.zfill(5)
    
    # Convertir columnas numéricas
    numeric_cols = ['geo_epgs_25831_x', 'geo_epgs_25831_y', 'geo_epgs_4326_lat', 'geo_epgs_4326_lon']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # Convertir IDs a tipo Int64 (permite NaN)
    id_cols = ['institution_id', 'addresses_roadtype_id', 'addresses_road_id', 
               'addresses_neighborhood_id', 'addresses_district_id', 
               'secondary_filters_id', 'secondary_filters_tree', 'secondary_filters_asia_id']
    for col in id_cols:
        df[col] = df[col].astype('Int64')
    
    # Limpiar strings solo en columnas de texto
    text_cols = ['name', 'institution_name', 'addresses_road_name', 
                 'addresses_neighborhood_name', 'addresses_district_name', 
                 'addresses_town', 'values_category', 'values_attribute_name',
                 'values_value', 'values_description', 'secondary_filters_name',
                 'secondary_filters_fullpath']
    
    for col in text_cols:
        mask = df[col].notna()  # Solo procesar valores no nulos
        if mask.any():
            df.loc[mask, col] = df.loc[mask, col].astype(str).str.strip()
    
    print("Tipos de datos actualizados correctamente")
    print("\nInformación del DataFrame:")
    print(df.info())
    
    # Mostrar algunos valores de ejemplo
    print("\nEjemplos de valores en columnas clave:")
    for col in ['created', 'addresses_zip_code', 'geo_epgs_4326_lat']:
        print(f"\n{col}:")
        print(df[col].head())
    
except Exception as e:
    print(f"Error durante la conversión: {str(e)}")
    print(f"Columna problemática: {col}")
    print(f"Tipo de datos actual: {df[col].dtype}")
    print(f"Primeros valores de la columna:")
    print(df[col].head())
```

```python
# Identificar columnas con pocos valores únicos o muchos nulos
def analizar_columnas_poco_informativas(df):
    # Contar valores únicos por columna
    nunique = df.nunique()
    # Contar valores no nulos por columna
    count_notnull = df.count()
    
    # Crear DataFrame con el análisis
    analisis = pd.DataFrame({
        'valores_unicos': nunique,
        'registros_no_nulos': count_notnull,
        'porcentaje_no_nulos': (count_notnull / len(df) * 100).round(2)
    }).sort_values('registros_no_nulos')
    
    print("Análisis de columnas:")
    print(analisis)
    
    # Identificar columnas a eliminar
    columnas_eliminar = []
    
    # Columnas con 0 o 1 valor único
    columnas_sin_variacion = nunique[nunique <= 1].index.tolist()
    columnas_eliminar.extend(columnas_sin_variacion)
    
    # Columnas con más de 90% de valores nulos
    columnas_muy_nulas = analisis[analisis['porcentaje_no_nulos'] < 10].index.tolist()
    columnas_eliminar.extend(columnas_muy_nulas)
    
    return list(set(columnas_eliminar))

# Identificar columnas a eliminar
columnas_eliminar = analizar_columnas_poco_informativas(df)

print("\nColumnas a eliminar:")
for col in columnas_eliminar:
    print(f"- {col}: {df[col].nunique()} valores únicos, {df[col].count()} registros no nulos")

# Crear nuevo DataFrame sin las columnas poco informativas
df_limpio = df.drop(columns=columnas_eliminar)

print("\nDimensiones originales:", df.shape)
print("Dimensiones después de limpieza:", df_limpio.shape)

# Mostrar columnas restantes
print("\nColumnas conservadas:")
for col in df_limpio.columns:
    print(f"- {col}")

# Guardar el DataFrame limpio
df = df_limpio.copy()
```

```python
# 1. Creación de categorías y flags adicionales
print("MEJORAS ADICIONALES Y CONTROL DE CALIDAD")
print("-" * 50)

# Crear DataFrame de trabajo
df_final = df.copy()

# Crear columna numero_direccion desde addresses_start_street_number
df_final['numero_direccion'] = df_final['addresses_start_street_number']

# 1. Categorización de direcciones
df_final['categoria_direccion'] = pd.cut(
    df_final['numero_direccion'],
    bins=[0, 50, 100, 200, 1050],
    labels=['Bajo', 'Medio', 'Alto', 'Muy alto']
)

# 2. Normalización temporal más detallada
df_final['decada'] = (df_final['año_creacion'] // 10) * 10
df_final['antiguedad'] = 2024 - df_final['año_creacion']
df_final['categoria_antiguedad'] = pd.cut(
    df_final['antiguedad'],
    bins=[0, 10, 20, 30, 40],
    labels=['Reciente', 'Establecido', 'Consolidado', 'Histórico']
)

# 3. Validación de valores atípicos
print("\n1. ANÁLISIS DE VALORES ATÍPICOS")
print("-" * 30)

def detectar_outliers(serie):
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return (serie < limite_inferior) | (serie > limite_superior)

# Detectar outliers en variables numéricas clave
variables_numericas = ['numero_direccion', 'año_creacion', 'geo_epgs_4326_lat', 'geo_epgs_4326_lon']
outliers = {}

for var in variables_numericas:
    outliers[var] = detectar_outliers(df_final[var])
    print(f"\nOutliers en {var}:")
    print(f"Cantidad: {outliers[var].sum()}")
    print(f"Porcentaje: {(outliers[var].sum() / len(df_final)) * 100:.2f}%")
    if outliers[var].sum() > 0:
        print("Valores atípicos:")
        print(df_final[outliers[var]][var].sort_values())

# 4. Análisis de rangos de dirección
df_final['tiene_rango'] = df_final['addresses_end_street_number'] != df_final['addresses_start_street_number']
df_final['diferencia_numeros'] = df_final['addresses_end_street_number'] - df_final['addresses_start_street_number']

print("\n2. ANÁLISIS DE RANGOS DE DIRECCIÓN")
print("-" * 30)
print(f"Direcciones con rango: {df_final['tiene_rango'].sum()}")
print("\nEstadísticas de diferencias:")
print(df_final[df_final['tiene_rango']]['diferencia_numeros'].describe())

# 5. Resumen de calidad de datos
print("\n3. RESUMEN DE CALIDAD DE DATOS")
print("-" * 30)

calidad = {
    'Completitud': {
        'Registros totales': len(df_final),
        'Registros con todos los datos': len(df_final.dropna()),
        'Porcentaje completitud': (len(df_final.dropna()) / len(df_final)) * 100
    },
    'Consistencia': {
        'Direcciones con rango': df_final['tiene_rango'].sum(),
        'Coordenadas válidas': len(df_final),
        'Años únicos': df_final['año_creacion'].nunique()
    },
    'Distribución temporal': {
        'Año más frecuente': df_final['año_creacion'].mode().iloc[0],
        'Rango de años': f"{df_final['año_creacion'].min()} - {df_final['año_creacion'].max()}",
        'Concentración en 1996': (df_final['año_creacion'] == 1996).mean() * 100
    }
}

for categoria, metricas in calidad.items():
    print(f"\n{categoria}:")
    for metrica, valor in metricas.items():
        print(f"- {metrica}: {valor}")

# 6. Estructura final
print("\n4. ESTRUCTURA FINAL DEL DATASET")
print("-" * 30)
print("\nColumnas finales:")
for col in df_final.columns:
    print(f"- {col}")
print(f"\nDimensiones finales: {df_final.shape}")

# 7. Recomendaciones
print("\n5. RECOMENDACIONES FINALES")
print("-" * 30)
recomendaciones = [
    "Investigar el pico de registros en 1996 (posible evento histórico)",
    "Validar los rangos de dirección superiores a 10 números",
    "Considerar una categorización más granular para períodos recientes",
    "Implementar validación automática de coordenadas para nuevos registros",
    "Documentar los criterios de categorización de direcciones"
]

for i, rec in enumerate(recomendaciones, 1):
    print(f"{i}. {rec}")
```
