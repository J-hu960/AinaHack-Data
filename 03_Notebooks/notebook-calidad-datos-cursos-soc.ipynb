```python
# 1. Convertir códigos a string manteniendo NaN
codigo_cols = ['codiarea', 'codimodalitat', 'codicentre']
for col in codigo_cols:
    df[col] = df[col].fillna('').astype(str)

# 2. Formatear códigos numéricos preservando NaN
def format_codigo(serie, digits=5):
    return serie.apply(lambda x: str(int(x)).zfill(digits) if pd.notnull(x) else '')

df['codimunicipi'] = format_codigo(df['codimunicipi'])
df['codipostal'] = format_codigo(df['codipostal'])

# 3. Limpiar campos numéricos
numeric_cols = ['duradacurs', 'importe', 'latitud', 'longitud']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 4. Convertir fecha
df['datainici'] = pd.to_datetime(df['datainici'], format='%d/%m/%Y', errors='coerce')

# 5. Limpiar campos de texto
text_cols = ['nomcurs', 'desccurscat', 'personesdest', 'requisits', 
             'temari', 'sortides', 'titulacio', 'nommodalitat',
             'financiacio', 'nomcentre', 'adrecacentre', 'telefon',
             'correu', 'web', 'departament', 'desccurses']

for col in text_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('').str.strip()

# 6. Eliminar columnas vacías
df = df.drop(['paraulesclau', 'georeferencia'], axis=1)

# 7. Verificar resultados
print("1. Tipos de datos actualizados:")
print(df.dtypes)

print("\n2. Valores nulos por columna:")
print(df.isnull().sum())

print("\n3. Estadísticas de campos numéricos:")
print(df[numeric_cols].describe().round(2))

print("\n4. Ejemplo de primeros registros de códigos:")
codigo_ejemplo = df[['codimunicipi', 'codipostal']].head()
print(codigo_ejemplo)
```

- Maneja los NaN en códigos
- Formatea códigos numéricos de forma segura
- Preserva los valores nulos donde sea apropiado
- Limpia y estandariza los campos de texto
- Muestra un resumen completo de la calidad de los datos
