```python
# 1. Optimización de tipos de datos
print("OPTIMIZACIÓN DE TIPOS DE DATOS")
print("-" * 50)

# Crear copia del DataFrame
df_optimizado = df.copy()

# 1. Conversión de fechas con formato ISO8601 y UTC
print("\nConvirtiendo fechas...")
for col in ['created', 'modified']:
    # Convertir a UTC y luego a datetime
    df_optimizado[col] = pd.to_datetime(df_optimizado[col], format='ISO8601', utc=True)
    
# Verificar la conversión
print("\nVerificación de tipos de fecha:")
print(df_optimizado['created'].dtype)

# 2. Conversión de IDs a Int64 (permite NaN)
columnas_id = [
    'institution_id', 'addresses_roadtype_id', 'addresses_road_id',
    'addresses_neighborhood_id', 'addresses_district_id',
    'values_id', 'values_attribute_id', 'secondary_filters_id',
    'secondary_filters_asia_id'
]

for col in columnas_id:
    df_optimizado[col] = df_optimizado[col].astype('Int64')

# 3. Conversión de números de calle a Int64
columnas_numeros = [
    'addresses_start_street_number', 'addresses_end_street_number',
    'addresses_zip_code'
]

for col in columnas_numeros:
    df_optimizado[col] = df_optimizado[col].astype('Int64')

# 4. Optimización de strings
columnas_string = [
    'register_id', 'name', 'institution_name', 'addresses_road_name',
    'addresses_neighborhood_name', 'addresses_district_name',
    'addresses_town', 'values_category', 'values_attribute_name',
    'values_value', 'values_outstanding', 'values_description',
    'secondary_filters_name', 'secondary_filters_fullpath'
]

for col in columnas_string:
    if df_optimizado[col].notna().any():  # Solo si hay valores no nulos
        df_optimizado[col] = df_optimizado[col].astype('string')

# 5. Añadir columnas derivadas de fecha
print("\nCreando columnas derivadas de fecha...")
df_optimizado['año_creacion'] = df_optimizado['created'].dt.year
df_optimizado['mes_creacion'] = df_optimizado['created'].dt.month
df_optimizado['dia_semana_creacion'] = df_optimizado['created'].dt.day_name()

# 6. Verificación de la optimización
print("\nCOMPARACIÓN DE USO DE MEMORIA")
print("-" * 30)
memoria_original = df.memory_usage(deep=True).sum() / 1024  # KB
memoria_optimizada = df_optimizado.memory_usage(deep=True).sum() / 1024  # KB
print(f"Memoria original: {memoria_original:.2f} KB")
print(f"Memoria optimizada: {memoria_optimizada:.2f} KB")
print(f"Reducción: {((memoria_original - memoria_optimizada) / memoria_original * 100):.2f}%")

# 7. Resumen de tipos de datos
print("\nTIPOS DE DATOS OPTIMIZADOS")
print("-" * 30)
print(df_optimizado.dtypes)

# 8. Verificación de integridad
print("\nVERIFICACIÓN DE INTEGRIDAD")
print("-" * 30)
print("\nConteo de valores nulos por columna:")
nulos = df_optimizado.isnull().sum()
print(nulos[nulos > 0])

# 9. Resumen temporal
print("\nRESUMEN TEMPORAL")
print("-" * 30)
print("\nRango de fechas:")
print(f"Primera fecha: {df_optimizado['created'].min()}")
print(f"Última fecha: {df_optimizado['created'].max()}")
print("\nDistribución por año:")
print(df_optimizado['año_creacion'].value_counts().sort_index())

# 10. Resumen final
print("\nRESUMEN FINAL")
print("-" * 30)
print(f"Dimensiones: {df_optimizado.shape}")
print(f"Columnas numéricas: {len(df_optimizado.select_dtypes(include=['int64', 'Int64', 'float64']).columns)}")
print(f"Columnas de texto: {len(df_optimizado.select_dtypes(include=['string', 'object']).columns)}")
print(f"Columnas de fecha: {len(df_optimizado.select_dtypes(include=['datetime64']).columns)}")
print(f"Columnas booleanas: {len(df_optimizado.select_dtypes(include=['bool']).columns)}")

# Guardar DataFrame optimizado
df = df_optimizado.copy()
```

```python
# 1. Limpieza adicional y creación de features relevantes
df_analisis = df.copy()

# Eliminar columnas con información redundante o poco relevante
columnas_a_eliminar_adicionales = [
    'values_category',  # solo contiene 'Telèfons'
    'addresses_town',   # solo 2 valores únicos
    'values_id', 'values_attribute_id', 'values_attribute_name',  # metadata de teléfonos
    'secondary_filters_id', 'secondary_filters_asia_id', 'secondary_filters_fullpath',  # redundantes
    'register_id'  # ID interno sin valor analítico
]

df_analisis = df_analisis.drop(columns=columnas_a_eliminar_adicionales)

# Crear categorías más amplias para tipos de centros
def categorizar_centro(tipo):
    if 'infantil' in tipo.lower():
        return 'Educación Infantil'
    elif 'primària' in tipo.lower():
        return 'Educación Primaria'
    elif 'secundària' in tipo.lower() or 'professional' in tipo.lower():
        return 'Educación Secundaria y FP'
    elif 'universitat' in tipo.lower():
        return 'Educación Superior'
    else:
        return 'Otros'

df_analisis['categoria_centro'] = df_analisis['secondary_filters_name'].apply(categorizar_centro)

# Crear períodos temporales
df_analisis['periodo_creacion'] = pd.cut(
    df_analisis['año_creacion'],
    bins=[1980, 1990, 2000, 2010, 2020, 2025],
    labels=['1980s', '1990s', '2000s', '2010s', '2020s']
)

print("ANÁLISIS MULTIDIMENSIONAL")
print("-" * 50)

# 1. Análisis por distrito y tipo de centro
print("\n1. DISTRIBUCIÓN DE CENTROS POR DISTRITO Y CATEGORÍA:")
dist_tipo = pd.crosstab(
    df_analisis['addresses_district_name'],
    df_analisis['categoria_centro'],
    margins=True
)
print(dist_tipo)

# 2. Evolución temporal
print("\n2. EVOLUCIÓN TEMPORAL POR CATEGORÍA:")
temp_tipo = pd.crosstab(
    df_analisis['periodo_creacion'],
    df_analisis['categoria_centro'],
    margins=True
)
print(temp_tipo)

# 3. Análisis de densidad educativa
print("\n3. INDICADORES POR DISTRITO:")
densidad = df_analisis.groupby('addresses_district_name').agg({
    'name': 'count',
    'categoria_centro': 'nunique',
    'año_creacion': ['min', 'max', 'mean']
}).round(2)

densidad.columns = ['total_centros', 'diversidad_tipos', 'año_primer_centro', 
                   'año_ultimo_centro', 'año_promedio']
print(densidad.sort_values('total_centros', ascending=False))

# 4. Análisis geográfico simplificado
print("\n4. DISTRIBUCIÓN GEOGRÁFICA:")
print("\nEstadísticas de ubicación por distrito:")
geo_stats = df_analisis.groupby('addresses_district_name').agg({
    'geo_epgs_4326_lat': ['mean', 'std'],
    'geo_epgs_4326_lon': ['mean', 'std']
}).round(4)
print(geo_stats)

# 5. Resumen temporal
print("\n5. RESUMEN TEMPORAL:")
print("\nCentros por década:")
decada = df_analisis['año_creacion'].apply(lambda x: (x//10)*10)
print(decada.value_counts().sort_index())

# 6. Métricas de diversidad
print("\n6. MÉTRICAS DE DIVERSIDAD EDUCATIVA:")
div_metrics = pd.DataFrame({
    'Total_Centros': df_analisis.groupby('addresses_district_name').size(),
    'Tipos_Diferentes': df_analisis.groupby('addresses_district_name')['categoria_centro'].nunique(),
    'Ratio_Diversidad': df_analisis.groupby('addresses_district_name')['categoria_centro'].nunique() / 
                       df_analisis.groupby('addresses_district_name').size()
}).round(3)
print(div_metrics.sort_values('Total_Centros', ascending=False))

# Guardar versión analítica
df = df_analisis.copy()
```

```python
# Análisis y tratamiento de nulos en variables numéricas
print("ANÁLISIS DE VARIABLES NUMÉRICAS")
print("-" * 50)

# 1. Análisis detallado de nulos
num_analysis = pd.DataFrame({
    'Nulos': num.isna().sum(),
    'Porcentaje': (num.isna().sum() / len(num) * 100).round(2),
    'Media': num.mean(),
    'Mediana': num.median(),
    'Desv_Std': num.std(),
    'Min': num.min(),
    'Max': num.max()
}).sort_values('Porcentaje', ascending=False)

print("\n1. ANÁLISIS DE NULOS NUMÉRICOS:")
print(num_analysis[num_analysis['Nulos'] > 0])

# 2. Estrategia de tratamiento por variable
estrategias = {
    'Eliminar': [
        'institution_id',           # 93.63% nulos, correlacionado con institution_name
        'addresses_end_street_number'  # 89.91% nulos, no esencial
    ],
    'Imputar': {
        'addresses_road_id': 'moda',           # 0.14% nulos
        'addresses_start_street_number': 'mediana',  # 0.14% nulos
        'addresses_neighborhood_id': 'moda',    # 0.14% nulos
        'addresses_district_id': 'moda',        # 0.14% nulos
        'addresses_zip_code': 'moda'            # 0.14% nulos
    }
}

# 3. Implementación de la limpieza
print("\n3. IMPLEMENTACIÓN DE LIMPIEZA:")

# Crear copia del DataFrame numérico
num_limpio = num.copy()

# Eliminar columnas con alto porcentaje de nulos
for col in estrategias['Eliminar']:
    num_limpio = num_limpio.drop(columns=col)
    print(f"Columna eliminada: {col}")

# Imputar valores según estrategia
for col, estrategia in estrategias['Imputar'].items():
    if estrategia == 'moda':
        valor = num_limpio[col].mode()[0]
    elif estrategia == 'mediana':
        valor = num_limpio[col].median()
    
    num_limpio[col] = num_limpio[col].fillna(valor)
    print(f"Columna {col} imputada con {estrategia}: {valor}")

# 4. Verificación post-limpieza
print("\n4. VERIFICACIÓN POST-LIMPIEZA:")
nulos_final = num_limpio.isna().sum()
print("\nNulos restantes:")
print(nulos_final[nulos_final > 0])

# 5. Análisis de distribución post-imputación
print("\n5. DISTRIBUCIÓN POST-IMPUTACIÓN:")
for col in estrategias['Imputar'].keys():
    print(f"\n{col}:")
    print(f"Antes - Desv. Std: {num[col].std():.2f}")
    print(f"Después - Desv. Std: {num_limpio[col].std():.2f}")
    print(f"Diferencia en la media: {(num_limpio[col].mean() - num[col].mean()):.2f}")

# 6. Validación de coherencia
print("\n6. VALIDACIÓN DE COHERENCIA:")
validaciones = {
    'Códigos postales válidos': (num_limpio['addresses_zip_code'] >= 8000).all(),
    'IDs de distrito válidos': (num_limpio['addresses_district_id'] > 0).all(),
    'Números de calle válidos': (num_limpio['addresses_start_street_number'] > 0).all()
}

for validacion, resultado in validaciones.items():
    print(f"{validacion}: {'✓' if resultado else '✗'}")

# Guardar DataFrame limpio
num = num_limpio.copy()
```

```python
# Mejoras y control de calidad
print("IMPLEMENTACIÓN DE MEJORAS Y CONTROL DE CALIDAD")
print("-" * 50)

# 1. Tratamiento de outliers
def tratar_outliers(df):
    df_limpio = df.copy()
    
    # 1.1 Número de dirección
    Q1 = df['addresses_start_street_number'].quantile(0.25)
    Q3 = df['addresses_start_street_number'].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5 * IQR
    
    # Crear categoría para outliers
    df_limpio['numero_direccion_cat'] = pd.cut(
        df_limpio['addresses_start_street_number'],
        bins=[-float('inf'), Q1, Q3, limite_superior, float('inf')],
        labels=['Bajo', 'Normal', 'Alto', 'Muy Alto']
    )
    
    # 1.2 Coordenadas
    # Crear bounds para Barcelona
    lat_bounds = (41.3, 41.5)
    lon_bounds = (2.0, 2.3)
    
    df_limpio['coordenadas_validas'] = (
        (df_limpio['geo_epgs_4326_lat'].between(*lat_bounds)) &
        (df_limpio['geo_epgs_4326_lon'].between(*lon_bounds))
    )
    
    return df_limpio

# 2. Validaciones adicionales
def validar_datos(df):
    validaciones = {
        'direcciones_validas': df['addresses_start_street_number'] > 0,
        'coordenadas_completas': df[['geo_epgs_4326_lat', 'geo_epgs_4326_lon']].notna().all(axis=1)
    }
    
    return pd.DataFrame(validaciones).all(axis=1)

# 3. Implementación de mejoras
df_mejorado = df.copy()
df_mejorado = tratar_outliers(df_mejorado)
df_mejorado['datos_validos'] = validar_datos(df_mejorado)

# 4. Resumen final
print("\nRESUMEN DE MEJORAS IMPLEMENTADAS:")
print("\n1. Distribución de categorías de número de dirección:")
print(df_mejorado['numero_direccion_cat'].value_counts(normalize=True).mul(100).round(2))

print("\n2. Validación de coordenadas:")
print(f"Coordenadas válidas: {df_mejorado['coordenadas_validas'].sum()}")
print(f"Porcentaje válido: {(df_mejorado['coordenadas_validas'].mean()*100):.2f}%")

print("\n3. Validación final de datos:")
print(f"Registros válidos: {df_mejorado['datos_validos'].sum()}")
print(f"Porcentaje válido: {(df_mejorado['datos_validos'].mean()*100):.2f}%")

# 5. Estadísticas de direcciones
print("\n4. ESTADÍSTICAS DE DIRECCIONES:")
stats_direcciones = {
    'Número mínimo': df_mejorado['addresses_start_street_number'].min(),
    'Número máximo': df_mejorado['addresses_start_street_number'].max(),
    'Media': df_mejorado['addresses_start_street_number'].mean(),
    'Mediana': df_mejorado['addresses_start_street_number'].median()
}
for key, value in stats_direcciones.items():
    print(f"{key}: {value:.2f}")

# 6. Recomendaciones de uso
recomendaciones = {
    'Filtrado': 'Usar datos_validos=True para análisis críticos',
    'Outliers': 'Considerar numero_direccion_cat para análisis espacial',
    'Geoespacial': 'Verificar coordenadas_validas para mapeos'
}

print("\n5. RECOMENDACIONES DE USO:")
for key, value in recomendaciones.items():
    print(f"{key}: {value}")

# Guardar versión mejorada
df = df_mejorado.copy()
```
